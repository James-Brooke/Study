```python
%matplotlib inline

import cv2
import numpy as np
import matplotlib.pyplot as plt

from keras import backend as K
from keras.applications import VGG16
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
```

```python
model = VGG16(weights='imagenet')
```

```python
img_path = r"C:\Users\James\Pictures\Weasel.jpg"

img = image.load_img(img_path, target_size=(224, 224))

x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
plt.imshow(img);
```

```python
preds = model.predict(x)
```

```python
decode_predictions(preds, 5)
```

```python
np.argmax(preds[0])
```

```python
weasel_output = model.output[:, 356]
last_conv_layer = model.get_layer('block5_conv3')
```

```python
grads = K.gradients(weasel_output, last_conv_layer.output)[0]
pooled_grads = K.mean(grads, axis=(0, 1, 2))
iterate = K.function([model.input], [pooled_grads, last_conv_layer.output[0]])
```

```python
pooled_grads_value, conv_layer_output_value = iterate([x])
```

```python
for i in range(512): #final layer of vgg16
    conv_layer_output_value[:, :, i] *= pooled_grads_value[i]
```

```python
heatmap = np.mean(conv_layer_output_value, axis=-1)
```

```python
plt.imshow(heatmap);
```

```python
heatmap = np.maximum(heatmap, 0)
heatmap /= np.max(heatmap)
plt.imshow(heatmap);
```

```python
opencvimg = cv2.imread(img_path)

heatmap = cv2.resize(heatmap, (opencvimg.shape[1], opencvimg.shape[0]))

heatmap = np.uint8(255 * heatmap)

heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

superimposed_img = heatmap * 0.8 + opencvimg
```

```python
cv2.imwrite(r"C:\Users\James\Pictures\Weaselheatmap.jpg", superimposed_img)
```
