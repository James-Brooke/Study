# IMDB movie reviews

```python
%matplotlib inline

import numpy as np

import matplotlib.pyplot as plt
from keras import models
from keras import layers

from keras.datasets import imdb
```

```python
(train_X, train_y), (test_X, test_y) = imdb.load_data(num_words=10000) #only 10000 most common words
```

```python
train_X.shape
```

```python
train_X[0] 
```

```python
index_dict = imdb.get_word_index()
reverse_index = {val: key for (key, val) in index_dict.items()}
```

```python
reverse_index
```

```python
translated = [reverse_index[i-3] if i > 2 else '' for i in train_X[0]] # keys 0-2 are reserved for padding
```

```python
' '.join(translated)
```

```python
train_y[0]
```

```python
def one_hotter(examples):
    out = np.zeros((len(examples), 10000)) #10000 words
    for i, example in enumerate(examples):
        out[i, example] = 1
    return out
```

```python
len(set(test_X[0])) #unique words in first example
```

```python
one_hotter(test_X)[0].sum() #checking one_hotter returns the same as above
```

```python
train_X = one_hotter(train_X)
test_X = one_hotter(test_X)
```

```python
train_X.dtype
```

```python
train_X = train_X.astype('float32')
test_X = test_X.astype('float32')
train_y = train_y.astype('float32')
test_y = test_y.astype('float32')
```

```python
val_X = train_X[:10000]
train_X = train_X[10000:]
val_y = train_y[:10000]
train_y = train_y[10000:]
```

```python
model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation ='sigmoid')) #Binary classification problem
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_X, train_y, epochs=20, batch_size=512, validation_data = (val_X, val_y))
```

```python
hist_dict = history.history
```

```python
hist_dict.keys()
```

```python
losses = hist_dict['loss']
val_losses = hist_dict['val_loss']
acc = hist_dict['acc']
val_acc = hist_dict['val_acc']
```

```python
plt.figure(figsize=(16,6))
plt.plot(range(1,21), losses, 'r', label='Train loss', linewidth=3)
plt.plot(range(1,21), val_losses, 'b', label='Validation loss', linewidth=3)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend();
```

```python
plt.figure(figsize=(16,6))
plt.plot(range(1,21), acc, 'r', label='Train accuracy', linewidth=3)
plt.plot(range(1,21), val_acc, 'b', label='Validation accuracy', linewidth=3)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend();
```

Clear overfitting.

```python
test = [[index_dict[i] for i in 'this movie sucks'.split()]]
test
```

```python
one_hotter(test)
```

```python
model.predict(one_hotter(test))
```

The model predicts 'this movie sucks' is a bad review, but it is not confident about it. 

```python
def tester(sentence):
    return model.predict(one_hotter([[index_dict[i] for i in sentence.split()]]))
```

```python
tests = ['this movie rules', 'it was ok', 'never again', 'two stars', 'worst movie i have ever seen'] 
```

```python
for i in tests:
    print('"{}"'.format(i), 'Prediction:', tester(i))
```

Doesn't seem to do well on short negative reviews.
