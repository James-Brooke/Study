# Feature maps

Convolutional neural networks make use of filters that create feature maps. 

Here a horizontal line and vertical line filter are applied to two images. The filters are manually created; however, in a normal scenario, the best filters are learned through cross-validation.

```python
%matplotlib inline

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def filter_image(image, direction='vertical'):
    
    height, width, channels = image.shape
    im_reshape = image.reshape(1, height, width, channels)
    
    filters = np.zeros(shape=(7,7,3,1), dtype=np.float32)
    
    if direction == 'vertical':
        filters[:, 3, :, :] = 1
    else:
        filters[3, :, :, :] = 1 #horizontal
        
    X = tf.placeholder(tf.float32, shape=(None, height, width, channels))
    conv = tf.nn.conv2d(X, filters, strides =[1,2,2,1], padding='SAME')
    
    with tf.Session() as sess:
        output = sess.run(conv, feed_dict={X:im_reshape})
        
    return output[0,:,:,0]
```

```python
cat = mpimg.imread(r"C:\Users\James\Pictures\Cat.jpg")
dog = mpimg.imread(r"C:\Users\James\Pictures\dog.jpeg")
```

```python
plt.imshow(cat);
```

### Vertical line filter

```python
plt.imshow(filter_image(cat), cmap='gray');
```

### Horizontal line filter

```python
plt.imshow(filter_image(cat, direction='horizontal'), cmap='gray');
```

```python
plt.imshow(dog)
```

### Vertical line filter

```python
plt.imshow(filter_image(dog), cmap='gray');
```

### Horizontal line filter

```python
plt.imshow(filter_image(dog, direction='horizontal'), cmap='gray');
```
